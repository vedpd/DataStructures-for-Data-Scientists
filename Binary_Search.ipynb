# Binary Search

1. Binary search works on the concept of searching an element in a list
2. Mandatory condition for Binary search is to have a sorted list (which is ascending sorted)

```
#Adding function to run binary search
def binSearch(A,key): #A is a list of elements
  l=0
  h=len(A)

  while(l<=h):
    mid=(l+h)//2

    if(key==A[mid]):
      return mid
    
    if(key<A[mid]):
      h=mid-1
    
    if(key>A[mid]):
      l= mid+1
  
  return 0
```

# **Validating case to find position of an element**

```
a=[2,5,8,12,16,23,38,56,72,91]

key=23
```

```
binSearch(a,key)
```
