## **Queue Attributes and methods**

1. Queue is ordered collection of items 
2. Addition happens at one end = <b>rear end -> Enqueue </b>
3. <b>Removal happens at the front end -> Dequeue </b>
4. This concept of removal and addition is known as <b> FIFO </b> (First In First Out)
3. Important methods are. :- <br>
  a. enqueue(item) - add element to the queue at the rear. Parameter: element to be added. Return: none <br>
  b. dequeue() - remove element from the front. Parameter : none. Return: item at front <br>
  c. isEmpty() - to check if queue is empty. Parametere : none. Return: boolean value <br>
  d. size() - returns the number of items in the queue. Parameter: None. Return: Integer <br>
  e. Queue(): creates a new queue which is empty. Parameters : None. Return: empty queue()


# **Queue Implementation**

```
class Queue():
  def __init__(self):
    self.items=[]
  
  def enqueue(self,item):
     return self.items.append(item) # add items from left to right, with left being first item/oldest

  def dequeue(self):
    return self.items.pop(0) # remove items at 0th index or left most- oldest thus following FIFO
  
  def isEmpty(self):
    return self.items ==[]
  
  def size(self):
    return len(self.items)
```
