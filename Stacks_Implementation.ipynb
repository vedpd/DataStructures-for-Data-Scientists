# Stack Concept and functions

1. Stack is ordered collection of items which are <b>added and removed from top </b>
2. This concept of removal and addition is known as <b>LIFO</b> (Last In First Out)
3. Important methods are. :- <br>
  a. push() - add element to the stack . Parameter: element to be pushed. Return: none <br>
  b. pop() - remove element from the top. Parameter : none. Return: item at top <br>
  c. peek() - look at item at top but do not remove it. Parameter : none. Return: item at top <br>
  d. isEmpty() - to check if stack is empty. Parametere : none. Return: boolean value <br>
  e. size() - returns the number of items in the stack. Parameter: None. Return: Integer <br>
  f. stack(): creates a new stack which is empty. Parameters : None. Return: empty stack()


## **Stack Implementation**

```
class Stack(object):
  
  def __init__(self):
    self.items = []

  def isEmpty(self):
    return self.items == [] # return comparison between self.items to an empty list : o/p = 0 or 1  ( yes or no)

  def push(self,item):
    return self.items.append(item)
  
  def pop(self):
    return self.items.pop()
  
  def peek(self):
    return self.items[len(self.items)-1] # total length of items in list = say 6, so index would be : 0,1,2,4,5.
    # last element is what should be returned and hence look at index = 5, which = len of list-1

  def size(self):
    return len(self.items)
```
